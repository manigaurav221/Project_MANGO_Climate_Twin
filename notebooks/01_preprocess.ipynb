```python
rain = pd.read_csv(
    "../data/raw/Rainfall_ind2025_rfp25.csv",
    sep=r"\s+",
    header=None
)

rain.head()
```

```python
date_rows = rain[
    rain[0] > 1000000
].index

date_rows[:10]
```

```python
len(date_rows)
```

```python
date_rows[:10].to_list()
```

```python
all_days = []
```

```python
for i in range(len(date_rows)):

    start = date_rows[i]

    if i < len(date_rows)-1:
        end = date_rows[i+1]
    else:
        end = len(rain)

    block = rain.iloc[start:end]

    date = str(int(block.iloc[0,0]))

    longitudes = block.iloc[0,1:].values

    grid = block.iloc[1:].copy()

    grid.columns = ["latitude"] + list(longitudes)

    grid = grid.apply(
        pd.to_numeric,
        errors="coerce"
    )

    grid = grid.replace(
        -999,
        np.nan
    )

    day = grid.melt(
        id_vars="latitude",
        var_name="longitude",
        value_name="rainfall"
    )

    day["date"] = date

    all_days.append(day)
```

```python
rain_final = pd.concat(
    all_days,
    ignore_index=True
)

rain_final.head()
```

```python
rain_final = rain_final.dropna(
    subset=["rainfall"]
)
```

```python
rain_final.head()
```

```python
rain_final.shape
```

```python
rain_final["date"] = (
    rain_final["date"]
    .astype(str)
    .str.zfill(8)
)

rain_final.head()
```

```python
rain_final["date"] = pd.to_datetime(
    rain_final["date"],
    format="%d%m%Y"
)

rain_final.head()
```

```python
rain_final["year"] = (
    rain_final["date"]
    .dt.year
)

rain_final["month"] = (
    rain_final["date"]
    .dt.month
)

rain_final["day"] = (
    rain_final["date"]
    .dt.day
)

rain_final.head()
```

```python
def rain_category(rain):
    if rain == 0:
        return "No Rain"

    elif rain<10 :
        return "Light"

    elif rain<50 :
        return "Moderate"

    else:
        return "Heavy"

rain_final["rain_category"] = (rain_final["rainfall"].apply(rain_category))
rain_final.head()
                               
```

## MaxTemp — clean pipeline

Same approach as rainfall: replace the missing sentinel (`99.90`), melt to long format, and **drop** out-of-domain cells instead of interpolating edges.

The raw file has **31 latitude rows × 365 days**. The first and last row in each day block are empty separators — they are removed automatically by `dropna`.

```python
MAXTEMP_SENTINEL = 99.90
ROWS_PER_DAY = 31
N_DAYS = 365  # align with rainfall (2013)

longitudes = np.arange(67.5, 98.5, 1)
latitudes = np.arange(7.5, 38.5, 1)

maxtemp_raw = pd.read_csv(
    "../data/raw/Maxtemp_MaxT_2025.CSV",
    sep=r"\s+",
    header=None,
)

maxtemp_raw = maxtemp_raw.replace(MAXTEMP_SENTINEL, np.nan)

print("Raw shape:", maxtemp_raw.shape)
print("Expected rows:", ROWS_PER_DAY * N_DAYS)
```

```python
all_maxtemp_days = []
dates = pd.date_range("2013-01-01", periods=N_DAYS, freq="D")

for day_idx in range(N_DAYS):
    start = day_idx * ROWS_PER_DAY
    end = start + ROWS_PER_DAY

    grid = maxtemp_raw.iloc[start:end].copy()
    grid.columns = longitudes
    grid["latitude"] = latitudes

    day = grid.melt(
        id_vars="latitude",
        var_name="longitude",
        value_name="max_temp",
    )

    day["date"] = dates[day_idx]
    all_maxtemp_days.append(day)

maxtemp_final = pd.concat(all_maxtemp_days, ignore_index=True)

maxtemp_final = maxtemp_final.dropna(subset=["max_temp"])

maxtemp_final.head()
```

```python
maxtemp_final["year"] = maxtemp_final["date"].dt.year
maxtemp_final["month"] = maxtemp_final["date"].dt.month
maxtemp_final["day"] = maxtemp_final["date"].dt.day

print("Shape:", maxtemp_final.shape)
print(
    "Latitude:", maxtemp_final["latitude"].min(), "-", maxtemp_final["latitude"].max()
)
print(
    "Longitude:", maxtemp_final["longitude"].min(), "-", maxtemp_final["longitude"].max()
)
print("Unique grid cells:", maxtemp_final.groupby(["latitude", "longitude"]).ngroups)
print("Unique dates:", maxtemp_final["date"].nunique())

maxtemp_final.describe()
```

```python
maxtemp_final.to_csv(
    "../data/processed/maxtemp_final.csv",
    index=False,
)

print("Saved maxtemp_final.csv")
```

## MinTemp — clean pipeline

Identical structure to max temp: replace `99.90`, melt, and drop out-of-domain cells.

```python
MINTEMP_SENTINEL = 99.90

mintemp_raw = pd.read_csv(
    "../data/raw/Mintemp_MinT_2025.CSV",
    sep=r"\s+",
    header=None,
)

mintemp_raw = mintemp_raw.replace(MINTEMP_SENTINEL, np.nan)

print("Raw shape:", mintemp_raw.shape)
print("Expected rows:", ROWS_PER_DAY * N_DAYS)
```

```python
all_mintemp_days = []

for day_idx in range(N_DAYS):
    start = day_idx * ROWS_PER_DAY
    end = start + ROWS_PER_DAY

    grid = mintemp_raw.iloc[start:end].copy()
    grid.columns = longitudes
    grid["latitude"] = latitudes

    day = grid.melt(
        id_vars="latitude",
        var_name="longitude",
        value_name="min_temp",
    )

    day["date"] = dates[day_idx]
    all_mintemp_days.append(day)

mintemp_final = pd.concat(all_mintemp_days, ignore_index=True)

mintemp_final = mintemp_final.dropna(subset=["min_temp"])

mintemp_final.head()
```

```python
mintemp_final["year"] = mintemp_final["date"].dt.year
mintemp_final["month"] = mintemp_final["date"].dt.month
mintemp_final["day"] = mintemp_final["date"].dt.day

print("Shape:", mintemp_final.shape)
print(
    "Latitude:", mintemp_final["latitude"].min(), "-", mintemp_final["latitude"].max()
)
print(
    "Longitude:", mintemp_final["longitude"].min(), "-", mintemp_final["longitude"].max()
)
print("Unique grid cells:", mintemp_final.groupby(["latitude", "longitude"]).ngroups)
print("Unique dates:", mintemp_final["date"].nunique())

mintemp_final.describe()
```

### Merge rainfall + max temp + min temp

Rainfall is on a **0.25°** grid; temperature is on a **1°** grid (coordinates at `.5`).
Snap rainfall coordinates to the nearest temp cell before joining.

```python
rain_for_merge = pd.read_csv(
    "../data/processed/rainfall_final.csv",
    parse_dates=["date"],
)

rain_for_merge["lat_match"] = np.floor(rain_for_merge["latitude"]) + 0.5
rain_for_merge["lon_match"] = np.floor(rain_for_merge["longitude"]) + 0.5

maxtemp_for_merge = maxtemp_final[
    ["date", "latitude", "longitude", "max_temp"]
].rename(
    columns={"latitude": "lat_match", "longitude": "lon_match"}
)

mintemp_for_merge = mintemp_final[
    ["date", "latitude", "longitude", "min_temp"]
].rename(
    columns={"latitude": "lat_match", "longitude": "lon_match"}
)

climate_merged = (
    rain_for_merge
    .merge(maxtemp_for_merge, on=["date", "lat_match", "lon_match"], how="inner")
    .merge(mintemp_for_merge, on=["date", "lat_match", "lon_match"], how="inner")
)

print("Merged rows:", len(climate_merged))
print(
    "Coverage:",
    f"{100 * len(climate_merged) / len(rain_for_merge):.1f}% of rainfall rows",
)

climate_merged[
    ["date", "latitude", "longitude", "rainfall", "max_temp", "min_temp"]
].head()
```

```python
climate_merged.to_csv(
    "../data/processed/climate_merged.csv",
    index=False,
)

print("Saved climate_merged.csv")
```

```python
temp_diff = (
    df.groupby(["latitude","longitude"])
    .agg({
        "max_temp":"mean",
        "min_temp":"mean"
    })
)

temp_diff["diff"] = (
    temp_diff["max_temp"]
    - temp_diff["min_temp"]
)

grid = temp_diff["diff"].unstack()

plt.figure(figsize=(12,10))

plt.pcolormesh(
    grid.columns,
    grid.index,
    grid.values,
    shading="auto",
    cmap="inferno"
)

plt.colorbar(label="Temperature Difference")

plt.title("Average Temperature Range Across India")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.show()
```
